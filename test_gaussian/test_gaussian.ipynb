```python
from scipy.ndimage.filters import gaussian_filter 
import numpy as np
```

```python
test_matrix = np.zeros((128,128))
test_matrix[50][50] = 1
```

```python
test_filtered = gaussian_filter(test_matrix,15)
for i in range(128):
    for j in range(128):
        if test_filtered[i][j] > 0.001:
            print(test_filtered[i][j])
test_filtered
```

```python
test_matrix = np.zeros((128,128))
test_matrix[50][50] = 1
test_matrix[60][60] = 1
```

```python
test_filtered = gaussian_filter(test_matrix,15)
for i in range(128):
    for j in range(128):
        if test_filtered[i][j] is 1:
            print(i,j)
test_filtered
```
